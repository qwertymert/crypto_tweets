```python
import pandas as pd
from combine_data import create_tweet_df

# pandas dataframe printing options for better analyzing
pd.set_option("display.min_rows", 40)  # min amount of rows to display
pd.set_option("display.max_rows", 100)  # max amount of rows to display before truncating
pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
pd.set_option("display.max_colwidth", None)

tweet_df = pd.read_csv("elonmusk_tweets.csv")  # read all tweets from csv file

price_df = pd.read_csv("btc_usdt_prices.csv")  # read all bitcoin prices since 2020-01-01
```

```python
custom_dataframe = create_tweet_df(tweet_df, price_df, likes=50000)
custom_dataframe
```

```python
# Customize dataframe by changing values
likes=25000  # tweets above N likes
keywords="coin|btc|doge|eth|crypto|chain|hodl|defi"  # keywords in tweets
extra_tweet_columns=["likes_count", "retweets_count", "link"]  # extra tweet data
change_price_interval=["15min", "2hr"]  # different time intervals for prices
percentage=True  # Calculate percentage changes
sort_by_percentage="hr"  # Sort by "hr"/"min"/"day"/None percentage change
sort_by_likes=False  #  Sort by likes (doesn't sort if sorted by percentage)
ascending=False  #  Sort by ascending order : True/False

df1 = create_tweet_df(tweet_df, price_df,
                      likes=likes,
                      keywords=keywords,
                      extra_tweet_columns=extra_tweet_columns,
                      change_price_interval=change_price_interval,
                      percentage=percentage,
                      sort_by_percentage=sort_by_percentage,
                      sort_by_likes=sort_by_likes,
                      ascending=ascending)

df1
```

```python
price_df2 = pd.read_csv("doge_usdt_prices.csv")
df2 = create_tweet_df(tweet_df, price_df2,
                      keywords="doge|dogecoin|dog|shib|inu|dogefather",
                      percentage=True,
                      sort_by_percentage="min",
                      change_price_interval=["15min"])
df2
```
